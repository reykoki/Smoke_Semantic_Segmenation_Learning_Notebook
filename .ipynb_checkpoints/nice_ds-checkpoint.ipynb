```python
import pickle
import random
import shutil
import os
import glob
import sys
from IPython.display import clear_output
sys.path.insert(1, './scripts')
sys.path.insert(1, './deep_learning')
from plot_tifs import plot_RGB,  plot_densities_from_processed_data, plot_R_G_B, plot_labels # code located in /scripts/plot_tifs.py


with open('/projects/mecr8410/semantic_segmentation_smoke/scripts/deep_learning/data_dict_new.pkl', 'rb') as handle:
    data_dict = pickle.load(handle)


    
```

```python
num_samples = [100, 10, 10]
data_types = ['train', 'val', 'test']
dst_root_dir = '/scratch/alpine/mecr8410/semantic_segmentation_smoke/sample_data/'

for idx, cat in enumerate(data_types):
    src_truth_fns = []
    truth_fns = random.choices(data_dict[cat]['truth'], k=num_samples[idx]*10)
    for truth_fn in truth_fns:
        fn = truth_fn.split('/')[-1]
        print(fn)
        plot_densities_from_processed_data(fn, data_loc="/scratch/alpine/mecr8410/semantic_segmentation_smoke/new_data/")
        x = input('yes or no')
        clear_output()
        if x == 'y':
            src_truth_fns.append(truth_fn)
        if len(src_truth_fns) >= num_samples[idx]:
            print('going to break')
            break
    
    src_data_fns = [s.replace('truth', 'data') for s in src_truth_fns]
    src_coords_fns = [s.replace('truth', 'coords') for s in src_truth_fns]
    dst_truth_fns = [s.replace('new_data', 'sample_data') for s in src_truth_fns]
    dst_data_fns= [s.replace('new_data', 'sample_data') for s in src_data_fns]
    dst_coords_fns= [s.replace('new_data', 'sample_data') for s in src_coords_fns]
    for fn_idx, fn in enumerate(src_truth_fns):
        shutil.copy2(src_coords_fns[fn_idx], dst_coords_fns[fn_idx])
        shutil.copy2(src_truth_fns[fn_idx], dst_truth_fns[fn_idx])
        shutil.copy2(src_data_fns[fn_idx], dst_data_fns[fn_idx])
```
